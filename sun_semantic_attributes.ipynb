```python
sun_att = loadmat('att_splits.mat')
```

```python
sun_att.keys()
```

```python
att_names = np.array(sun_att['allclasses_names'])
att_names = [a[0] for a in att_names]
```

```python
equivalents = {'ballroom':['ballroom'], 'bar_lounge': ['bar'], 'bathroom':['bathroom'], 'beach':['beach'], 'breakfast':['hotel_breakfast_area'], 'city_view':['city', 'downtown', 'street', 'tower'], 
              'golf_course':['golf_course'], 'guest_room':['bedroom'], 'health_club':['gymnasium_indoor'], 'hotel_front':['hotel_outdoor','factory_outdoor','skyscraper'],
              'kitchen':['kitchen','kitchenette'], 'living_room':['living_room'], 'lobby_view':['lobby','landing','reception'], 'meeting_room':['auditorium'], 'natural_view':['mountain','mountain_path', 'mountain_road','mountain_snowy'],
              'pool_view':['swimming_pool_outdoor'], 'restaurant':['restaurant']}
```

```python
categories = list(equivalents.keys())
```

```python
len(equivalents.keys())
```

```python
attributes = sun_att['att']
```

```python
import numpy as np
import pandas as pd
```

```python
attributes = np.array(attributes)
```

```python
attributes = attributes.T
```

```python
attributes.shape
```

```python
final_att = []
for k in equivalents.keys():
    l = equivalents[k]
    t = np.isin(att_names, l)
    att = attributes[t]
    mean = att.mean(axis=0)
    final_att.append(mean)
    
```

```python
final_attributes = np.array(final_att)
```

```python
final_attributes.shape
```

```python
semantic = pd.read_pickle('semantic_embeddings.pkl')
```

```python
df = pd.DataFrame(final_attributes)
```

```python
df['Class'] = categories
```

```python
df.to_pickle('sun_attributes.pkl')
```
